```python
df.shape
```

# SENTIMENT ANALYSIS

```python
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
df["sentiment"] = df["comment"].apply(lambda x: sentiment_pipeline(x)[0]["label"])
df.head()
```

# Generate Summary of Comments

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

def generate_summary(text):
    try:
        summary = summarizer(text, max_length=9, min_length=5, do_sample=False)
        return summary[0]['summary_text']
    except:
        return text

df["summary"] = df["comment"].apply(generate_summary)

df.head()
```

```python
df_sample = df.head(5)
df_sample["summary"] = df_sample["comment"].apply(generate_summary)
df_sample
```
